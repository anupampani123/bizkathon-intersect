```python
# we try to allocate portfolio efficiently based on sharpe ratio
#goal is to assign weights to the assets to maximize returns for a given level of volatility 
#  efficient frontier - uses possible combination of riskey assets which ca be plotted in the risk expected return space
#upper edge of the hyperbola obtained is the efficient frontier in the absence of a risk free asset
#The line running tangent to the hyperbola represents the Capital Allocation Line, whose slope measures the additional return 
#one receives from taking on more risk



alist = [(x+1) for x in range(len(industry))]
w0 = [x / np.sum(alist) for x in alist]

# create df and portfolio
industry_data = get_data(key, industry)
industry_df = create_df(industry_data, start_date)


```
